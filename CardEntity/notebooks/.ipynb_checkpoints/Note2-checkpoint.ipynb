```python
import numpy as np
import pandas as pd
import string
import re
import os
```

```python
df = pd.read_excel('../dataset/raw/dataset_raw3_final1_esti.xlsx')
# df = pd.read_csv('../dataset/raw/dataset_raw.csv')
```

```python
df.head(10)
```

```python
whitespace = string.whitespace
punctuation = "!#$%&\'()*+:;<=>?[\\]^`{|}~"

tableWhitespace = str.maketrans('', '', whitespace)
tablePunctuation = str.maketrans('', '', punctuation)

def cleanText(txt):
    # Convert to string and lower case
    text = str(txt).lower()
    # Remove whitespace and punctuation
    removewhitespace = text.translate(tableWhitespace)
    removepunctuation = removewhitespace.translate(tablePunctuation)
    return removepunctuation

# Assume df is the DataFrame you are working with
df['text'] = df['text'].apply(cleanText)

# Create a copy with the query to avoid SettingWithCopyWarning
dataClean = df.query("text != '' ").copy()
dataClean.dropna(inplace=True)

# Menampilkan dataClean setelah proses pembersihan
dataClean
```

```python
# Convert SpaCy Format
```

```python
group = dataClean.groupby(by='id')
```

```python
cards = group.groups.keys()
len(cards)
```

```python
allCardsData = []
for card in cards:
    cardData = []
    grouparray = group.get_group(card)[['text','tag']].values
    content = ''
    annotations = {'entities':[]}
    start = 0
    end = 0
    for text, label in grouparray:
        text = str(text)
        stringLength = len(text) + 1

        start = end
        end = start + stringLength

        if label != 'O':
            annot = (start,end-1,label)
            annotations['entities'].append(annot)

        content = content + text + ' '
        
        
    cardData = (content,annotations)
    allCardsData.append(cardData)
```

```python
allCardsData
```

```python
## Split the Data into Training and Testing Set

import random
random.shuffle(allCardsData)
len(allCardsData)
```

```python
totalCards = len(allCardsData)
trainRatio = 0.8  # Persentase data latih
testRatio = 1 - trainRatio  # Persentase data uji

numTrainData = int(totalCards * trainRatio)
numTestData = totalCards - numTrainData

print(numTrainData)
print(numTestData)
TrainData = allCardsData[:numTrainData]
TestData = allCardsData[numTrainData:]
```

```python
import spacy
from spacy.tokens import DocBin
import pickle

nlp = spacy.blank("id")
```

```python
training_data = pickle.load(open('../dataset/pickle/TrainData.pickle','rb'))
testing_data = pickle.load(open('../dataset/pickle/TestData.pickle','rb'))
```

```python
db = DocBin()
for text, annotations in training_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations['entities']:
        span = doc.char_span(start, end, label=label)
        ents.append(span)
    doc.ents = ents
    db.add(doc)
db.to_disk("../dataset/model/train.spacy")
```

```python
# the DocBin will store the example documents
db_test = DocBin()
for text, annotations in testing_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations['entities']:
        span = doc.char_span(start, end, label=label)
        ents.append(span)
    doc.ents = ents
    db_test.add(doc)
db_test.to_disk("../dataset/model/test.spacy")
```

```python
import pytesseract
import cv2
```

```python
model_ner = spacy.load('../dataset/model/output/model-best')

image = cv2.imread('../dataset/img/test2.jpeg')
# image = cv2.imread('../dataset/card/card16.jpeg')

# extract data using Pytesseract 
tessData = pytesseract.image_to_data(image)
# convert into dataframe
tessList = list(map(lambda x:x.split('\t'), tessData.split('\n')))
df = pd.DataFrame(tessList[1:],columns=tessList[0])
df.dropna(inplace=True) # drop missing values
df['text'] = df['text'].apply(cleanText)

# convet data into content
df_clean = df.query('text != "" ')
content = " ".join([w for w in df_clean['text']])
print(content)

doc = model_ner(content)
from spacy import displacy
docjson = doc.to_json()
docjson.keys()
doc_text = docjson['text']
datafram_tokens = pd.DataFrame(docjson['tokens'])
datafram_tokens['token'] = datafram_tokens[['start','end']].apply(
    lambda x:doc_text[x[0]:x[1]] , axis = 1)
datafram_tokens.head(10)
right_table = pd.DataFrame(docjson['ents'])[['start','label']]
datafram_tokens = pd.merge(datafram_tokens,right_table,how='left',on='start')
datafram_tokens.fillna('O',inplace=True)
datafram_tokens.head(100)
```

```python
import json

# Mengonversi DataFrame menjadi format yang sesuai dengan struktur JSON yang diinginkan
datafram_tokens.iterrows()
result_json = []
for index, row in datafram_tokens.iterrows():
    if row["label"] != "O":
        result_json.append({
            "token": row["token"],
            "label": row["label"]
        })
result_json
```
